```python
import pandas as pd
df=pd.read_csv("Housing.csv")
df.shape
```

```python
print(df["mainroad"].unique())
print(df["guestroom"].unique())
print(df["basement"].unique())
print(df["hotwaterheating"].unique())
print(df["airconditioning"].unique())
print(df["prefarea"].unique())
print(df["furnishingstatus"].unique())
```

```python
df=pd.get_dummies(df,drop_first=True)
df.head()
```

```python
df.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
#get correlations of each features in dataset
corrmat = df.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20,20))
#plot heat map
g=sns.heatmap(df[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

```python
df[["mainroad_yes","guestroom_yes","basement_yes","hotwaterheating_yes","airconditioning_yes","prefarea_yes","furnishingstatus_semi-furnished","furnishingstatus_unfurnished"]]=df[["mainroad_yes","guestroom_yes","basement_yes","hotwaterheating_yes","airconditioning_yes","prefarea_yes","furnishingstatus_semi-furnished","furnishingstatus_unfurnished"]].astype(int)
df
```

```python
X=df.iloc[:,1:]
y=df.iloc[:,0]
X.head()
```

```python
y.head()
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
model = LinearRegression()
model.fit(X,y)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import RFE
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
model.fit(X_train,y_train)
rfe = RFE(estimator=model, n_features_to_select=10) # running RFE
rfe = rfe.fit(X_train, y_train)
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
#creating dataframe with important features
num_vars = X_train.columns[rfe.support_]
num_vars
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()

'''num_vars=['area', 'bedrooms', 'bathrooms', 'stories', 'parking', 'mainroad_yes',
       'guestroom_yes', 'basement_yes', 'hotwaterheating_yes',
       'airconditioning_yes', 'prefarea_yes',
       'furnishingstatus_semi-furnished', 'furnishingstatus_unfurnished']'''

X_train[num_vars]= scaler.fit_transform(X_train[num_vars])
X_train
```

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[num_vars]
# Adding a constant variable 
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
model = sm.OLS(y_train,X_train_rfe).fit()   # Running the linear model
model.summary()
X_train_rfe.dtypes
```

```python
#Dropping basement_yes as it is insignificant.
X_train1= X_train_rfe.drop(columns='basement_yes')
X_train1
```

```python
X_train2= sm.add_constant(X_train1)
model1 = sm.OLS(y_train,X_train2).fit() 
model1.summary()
X_train2
```

```python
X_train_new= X_train2.drop('const',axis=1)
model1 = sm.OLS(y_train,X_train_new).fit()
X_train_new.columns
```

```python
#residual analysis
y_train_pred = model1.predict(X_train_new)
fig = plt.figure()
sns.distplot((y_train - y_train_pred), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)     
#print(y_train_pred)
```
